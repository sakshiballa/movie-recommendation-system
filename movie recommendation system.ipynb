```python
df = df[['title', 'genres', 'director', 'cast', 'tagline']]
df = df.fillna('')
df['text'] = df['title'] + ' ' + df['genres'] + ' ' + df['director'] + ' ' + df['cast'] + ' ' + df['tagline']
import string
df['text'] = df['text'].apply(lambda x: x.lower().translate(str.maketrans('', '', string.punctuation)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])
```

```python
#algo
#1-find index of mov
#2-calc cosine similarity betwn movies selected
#3-sim scores in desc order except the movie in topn
#4-retrieve index of rec movies
#5-return recs
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(X)
def get_recommendations(title, cosine_sim=cosine_sim, df=df, top_n=10):
    idx = df[df['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    movie_indices = [i[0] for i in sim_scores]
    return df['title'].iloc[movie_indices]
```

```python
get_recommendations('The Godfather')
```

```python
get_recommendations("Avatar")
```
